# Config template insertion

```python
from emma.analyzer_utils import parse_cisco_config

with open("test_data/leaf-1.11545.conf") as f:
    config = f.read()

parsed = parse_cisco_config(config)

template = """{% for server in default_ntp_servers %}
ntp server {{ server }} use-vrf default
{% endfor %}"""
```

```python
"".join(parsed["ntp"]) in config
```

```python
templated = config.replace("".join(parsed["ntp"]), template)
```
